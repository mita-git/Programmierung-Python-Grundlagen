### Dictionarys {key1: value1, key2: value2}
siehe 1. Semester
Eine Möglichkeit, sie herzustellen, besteht im Einsatz von geschweiften Klammern {} und Doppelpunkten zum Trennen der Schlüssel von den Werten.<br>

````python
words = ['Apfel', 'Bad', 'Bar', 'Atom', 'Buch', 'Pferd', 'Hund', 'Pflaume']
by_letter = {}
for word in words:
    letter = word[0]
    if letter not in by_letter:
        by_letter[letter] = [word]
    else:
        by_letter[letter].append(word)
by_letter
````

#### setdefault
Die Methode setdefault dient genau diesem Zweck.<br>

````python
by_letter=dict()
for word in words:
    letter = word[0]
    by_letter.setdefault(letter, []).append(word)
by_letter
````

### Funktionen sind Objekte
Da Python-Funktionen Objekte sind, lassen sich viele in anderen Sprachen schwierige Konstrukte hier leicht ausdrücken.<br>

````python
states = ['   Alabama ', 'Georgia!', 'Georgia', 'georgia', 'FlOrIda', 'south   carolina##', 'West virginia?']
````

````python
import re

def clean_strings1(strings):
    result = []
    for value in strings:
        value = value.strip()
        value = re.sub('[!#?]', '', value)
        value = value.title()
        result.append(value)
    return result
````

````python
clean_strings1(states)
````

````python
def remove_punctuation(value):
    return re.sub('[!#?]', '', value)

clean_ops = [str.strip, remove_punctuation, str.title]

def clean_strings2(strings, ops):
    result = []
    for value in strings:
        for function in ops:
            value = function(value)
        result.append(value)
    return result
````

````python
clean_strings2(states, clean_ops)
````

### Currying: teilweise Anwendung von Argumenten

### Generatoren

Ein wichtiges Merkmal in Python ist eine konsistente Methode, über Sequenzen zu iterieren, etwa über Objekte in einer Liste oder Zeilen in einer Datei. Erreicht wird das durch das Iterator-Protokoll, eine generische Möglichkeit, Objekte iterierbar zu machen.

Um einen Generator herzustellen, nutzen Sie in einer Funktion das Schlüsselwort **yield** anstelle von return:

````python
def squares(n=10):
    print('Quadrate erzeugen aus 1 to {0}'.format(n ** 2))
    for i in range(1, n + 1):
        yield i ** 2
````

````python
gen = squares(7)
print(gen)
for x in gen:
    print(x, end=' ')
````

#### Generator-Ausdrücke

Eine noch knappere Methode, einen Generator herzustellen, bieten Generator-Ausdrücke. Dabei handelt es sich um einen Generator analog zu List, Dict und Set Comprehensions; um einen zu erzeugen, schließen Sie das, was ansonsten eine List Comprehension wäre, in runde statt in eckige Klammern ein:

````python
gen = (x ** 2 for x in range(100))
print(gen)
````

Das ist vollständig äquivalent zu dem folgenden deutlich ausführlicheren Generator:

````python

def _make_gen():
    for x in range(100):
        yield x ** 2
gen = _make_gen()
print(gen)
````

### Das Modul itertools

Das Standardbibliotheksmodul itertools besitzt eine Kollektion aus Generatoren für viele gebräuchliche Datenalgorithmen.

So nimmt zum Beispiel **groupby** eine Sequenz und eine Funktion entgegen.<br>
Der Rückgabewert der Funktion ist das Kriterium, nach dem aufeinanderfolgende Elemente in der Sequenz gruppiert werden.

````python
import itertools
first_letter = lambda x: x[0]
names = ['Alan', 'Adam', 'Wes', 'Will', 'Albert', 'Steven']
names.sort()
for letter, names in itertools.groupby(names, first_letter):
    print(letter, list(names)) # names is a generator
````
