## Try und except

```python
# Copyright (c) 2022, Michael Talarczyk
# Einfaches Programm zur Kreisberechnung

def Kreis (EingabeRadius=1):   # Aufrufen der Funktion mit Defaultwert=1
    # Hilfetext für die help() Funktion zwischen ''' ''' oder """  """
    ''' 
    Defaultwert für den Radius ist 1
    Rückgabe: Radius,Umfang,Fläche,Durchmesser
    '''
    PI = 3.14159265359
    Radius = float()
    try:                      # Wenn in dem im folgenden eine Fehler auftriit springe nach except:
        Radius = float(EingabeRadius)
        Umfang = 2 * PI * Radius
        Fläche = PI * Radius ** 2
        Durchmesser = 2 * Radius
        return (Radius,Umfang,Fläche,Durchmesser)     
    except:                   # Bei einem Fehler get es hier weiter
        print("Eine Fehler ist aufgetreten, die Eingabe kann nicht in eine Fließkommazahl umgewandelt werde.")
    
    print('weiter')
    
    
Kreis()
```

```python
# Die Funktion randint nimmt die Parameter low und high und gibt eine
# ganze Zahl zwischen low und high (einschließlich) zurück

import random
for i in range(10):
    x = random.randint(1, 6)
    print(x,end=';')
```

## while True: break und continue

```python
def durchschnitt():
    '''
    Sammeltwerte und gibt die Anzahl, Summe und den Durchschnitt zurück.
    Mit done die Eingabe beenden.
    '''
    Zeile = ""
    Zahl = 0
    anzahl = 0
    summe = 0
    durchschnitt = 0
    while True:
        Zeile = input('Bitte eine Zahl eingeben: ')
        if Zeile == 'done':
            break                                    # Bei Eingabe von done bricht die while Schleife ab
        try:
            Zahl = float(Zeile.replace(",", "."))    # replace ersetzt innerhalb eine Textes Zeichen
            summe = summe + Zahl
            anzahl = anzahl + 1
            continue                                 # Die while Schleife wird fortgesetzt
        except:
            print("Es werden nur Zahlen akzeptiert. Beenden mit 'done'")
    if anzahl==0:                                    # um die division durch 0 zu verhindern
        anzahl=1
    durchschnitt=summe/anzahl
    # print("Anzahl: ",anzahl," Summe: ",summe," Durchschnitt: ",durchschnitt)
    return (anzahl,summe,durchschnitt)
```
