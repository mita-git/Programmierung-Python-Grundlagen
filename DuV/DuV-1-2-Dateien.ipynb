Eine der Möglichkeiten, das Säubern geöffneter Dateien zu erleichtern, bietet die **with-Anweisung**

```python
path = "Das Leben ein Traum.txt" 
with open(path) as fh:
    lines = [x.rstrip() for x in fh]
lines
```

Dabei wird die Datei fh automatisch geschlossen, wenn der with-Block verlassen wird.

```python
def createTxt(path):
    try:
        f = open(path, 'xt', encoding='utf-8')
    except FileNotFoundError:
        print('FileNotFoundError', path)
    except FileExistsError:
        print('FileExistsError', path)
    except NameError:
        print('NameError', path)
    else:
        print("else: Datei", path,"erzeugt")
        f.write('Die ist ein äüöß Text')
    finally:
        print("finally: Close")
        try:
            f.close()
        except UnboundLocalError:
            print(path,"not open")
            
def showTxt_utf8(path):
    try:
        with open(path,encoding='utf-8') as f:
            linesList = [x.rstrip() for x in f]
        print(linesList)   
        
        f = open(path, 'r', encoding='utf-8')
        for line in f:
            print(line.rstrip())        

    except FileNotFoundError:
        print('FileNotFoundError', path)
    except UnicodeDecodeError:
        print('UnicodeDecodeError', path)
    finally:
        print("finally: Close")
        try:
            f.close()
        except UnboundLocalError:
            print(path,"not open")        
```

```python
Datei = 'examples/segismundo.txt'  #  FileNotFoundError
createTxt(Datei)
```
