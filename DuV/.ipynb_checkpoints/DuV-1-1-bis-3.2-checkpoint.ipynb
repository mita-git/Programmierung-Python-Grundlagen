### None

None ist der Wert null in Python. Liefert eine Funktion nicht explizit einen Wert zurck, liefert sie implizit None zurck.<br>
None ist darber hinaus ein gebruchlicher Standardwert fr Funktionsargumente<br>

### Dictionarys {key1: value1, key2: value2}
siehe 1. Semester
Eine Mglichkeit, sie herzustellen, besteht im Einsatz von geschweiften Klammern {} und Doppelpunkten zum Trennen der Schlssel von den Werten.<br>

````python
words = ['Apfel', 'Bad', 'Bar', 'Atom', 'Buch', 'Pferd', 'Hund', 'Pflaume']
by_letter = {}
for word in words:
    letter = word[0]
    if letter not in by_letter:
        by_letter[letter] = [word]
    else:
        by_letter[letter].append(word)
by_letter
````

#### setdefault
Die Methode setdefault dient genau diesem Zweck.<br>

````python
by_letter=dict()
for word in words:
    letter = word[0]
    by_letter.setdefault(letter, []).append(word)
by_letter
````

#### defaultdict
Das eingebaute collections-Modul besitzt eine ntzliche Klasse namens defaultdict, die dies noch weiter vereinfacht. <br>

````python
by_letter = {}
from collections import defaultdict
by_letter = defaultdict(list)
for word in words:
    by_letter[word[0]].append(word)
    
by_letter
````

````python
import re
def findHandyNr(dicNr, regEx):
    res = []
    for key in dicNr.keys():
#        print(key,regEx)
#        match = re.match(regEx,str(key))
#        print(match)
        if re.match(regEx,str(key)):
            res.append(dicNr[key])
#    print(len(res))
    if len(res) == 1:
        return res
    elif len(res) > 1:
        return "Mehrere Treffer"
    elif len(res) == 0:
        return "Keine Treffer"
    else:
        return "error"
````

### Funktionen sind Objekte
Da Python-Funktionen Objekte sind, lassen sich viele in anderen Sprachen schwierige Konstrukte hier leicht ausdrcken.<br>

````python
states = ['   Alabama ', 'Georgia!', 'Georgia', 'georgia', 'FlOrIda', 'south   carolina##', 'West virginia?']
````

````python
import re

def clean_strings1(strings):
    result = []
    for value in strings:
        value = value.strip()
        value = re.sub('[!#?]', '', value)
        value = value.title()
        result.append(value)
    return result
````

````python
clean_strings1(states)
````

````python
def remove_punctuation(value):
    return re.sub('[!#?]', '', value)

clean_ops = [str.strip, remove_punctuation, str.title]

def clean_strings2(strings, ops):
    result = []
    for value in strings:
        for function in ops:
            value = function(value)
        result.append(value)
    return result
````

````python
clean_strings2(states, clean_ops)
````

### Currying: teilweise Anwendung von Argumenten
Currying ist Informatiker-Slang (nach dem Mathematiker Haskell Curry) und bedeutet, dass man neue aus existierenden Funktionen ableitet durch die teilweise Anwendung von Argumenten. 

````python
def add_numbers(x, y):
    return x + y
````

````python
add_five = lambda y: add_numbers(5, y)
add_five(37)
````

Das eingebaute functools-Modul kann diesen Vorgang mit der partial-Funktion vereinfachen

````python
from functools import partial
add_five = partial(add_numbers, 5)
add_five(42)
````

### Generatoren

Ein wichtiges Merkmal in Python ist eine konsistente Methode, ber Sequenzen zu iterieren, etwa ber Objekte in einer Liste oder Zeilen in einer Datei. Erreicht wird das durch das Iterator-Protokoll, eine generische Mglichkeit, Objekte iterierbar zu machen.

````python
# Zum Beispiel erhlt man beim Iterieren ber ein Dictionary dessen Schlssel
some_dict = {'a': 1, 'b': 2, 'c': 3}
for key in some_dict:
    print("Key",key)
````

````python
dict_iterator = iter(some_dict)
print(dict_iterator)
````

````python
list(dict_iterator)
````

Um einen Generator herzustellen, nutzen Sie in einer Funktion das Schlsselwort **yield** anstelle von return:

````python
def squares(n=10):
    print('Quadrate erzeugen aus 1 to {0}'.format(n ** 2))
    for i in range(1, n + 1):
        yield i ** 2
````

````python
gen = squares(7)
print(gen)
for x in gen:
    print(x, end=' ')
````

#### Generator-Ausdrcke

Eine noch knappere Methode, einen Generator herzustellen, bieten Generator-Ausdrcke. Dabei handelt es sich um einen Generator analog zu List, Dict und Set Comprehensions; um einen zu erzeugen, schlieen Sie das, was ansonsten eine List Comprehension wre, in runde statt in eckige Klammern ein:

````python
gen = (x ** 2 for x in range(100))
print(gen)
````

Das ist vollstndig quivalent zu dem folgenden deutlich ausfhrlicheren Generator:

````python

def _make_gen():
    for x in range(100):
        yield x ** 2
gen = _make_gen()
print(gen)
````

### Das Modul itertools

Das Standardbibliotheksmodul itertools besitzt eine Kollektion aus Generatoren fr viele gebruchliche Datenalgorithmen.

So nimmt zum Beispiel **groupby** eine Sequenz und eine Funktion entgegen.<br>
Der Rckgabewert der Funktion ist das Kriterium, nach dem aufeinanderfolgende Elemente in der Sequenz gruppiert werden.

````python
import itertools
first_letter = lambda x: x[0]
names = ['Alan', 'Adam', 'Wes', 'Will', 'Albert', 'Steven']
names.sort()
for letter, names in itertools.groupby(names, first_letter):
    print(letter, list(names)) # names is a generator
````
