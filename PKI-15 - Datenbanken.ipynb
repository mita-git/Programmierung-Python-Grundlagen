## Datenbankkonzepte


Auf den ersten Blick sieht eine Datenbank aus wie eine Tabellenkalkulation mit mehreren Sheets.<br>
Die primären Datenstrukturen in einer Datenbank sind: Tabellen, Zeilen und Spalten.<br>

```python
import sqlite3

conn = sqlite3.connect('music.sqlite')
cur = conn.cursor()

cur.execute('DROP TABLE IF EXISTS Tracks')         # DROP Damit in dem Beispielen immer eine leere Daten Bank ist
cur.execute('CREATE TABLE Tracks (title TEXT, plays INTEGER)')    # Anlegen der Tabelle

# Spalte title vom Type Text und Spalte plays vom Type Integer

conn.close()
```

## Datenbankbrowser für SQLite

**CURSOR** ist wie der Filehandler bei Dateien zu verstehen

```python
import sqlite3

conn = sqlite3.connect('music.sqlite')
cur = conn.cursor()

cur.execute('INSERT INTO Tracks (title, plays) VALUES (?, ?)', ('Thunderstruck', 20))   # eintagen von Daten
cur.execute('INSERT INTO Tracks (title, plays) VALUES (?, ?)', ('My Way', 15))          # eintagen von Daten
conn.commit()

print('Tracks:')
cur.execute('SELECT title, plays FROM Tracks')                                        # Daten auslesen
for row in cur:
     print(row)

```

```python
cur.execute('DELETE FROM Tracks WHERE plays < 100')                                   # Daten wieder löschen
conn.commit()

cur.close()
```
