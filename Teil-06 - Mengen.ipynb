### ohne (mengentheoretische Differenz)




$ \mathbb{N} = \{0,1,2,3,4,5,6,....\}$ Menge der natürliche Zahlen

$ A = \{n : n \in \mathbb{N} \ und \ n \leqslant 100\}$

```python
A = set(range(101))
print(A)
```

```python
# D besteht also aus allen Elementen von A, die kleiner als 42 und nicht durch 3 teilbar sind.
D = set()
for n in A:
    if n < 42 and n % 3 != 0:
        D.add(n)
print(D)
```

```python
E3 = {n for n in A if n < 42 and n % 3 != 0} # auch in einer Zeile
print (E3)
```

```python
A = set(range(101))
Q1 = {m ** 2 for m in A if m ** 2 < 500}       # Menge der Quadratzahlen    
print(Q1)
```

# Darstellung vom Brüchen in Python
## modul fractions = Brüche

```python
from fractions import Fraction as Bruch                 # fractions = Brüche

Zähler = {1,2,5}
Nenner = {3,7,9,11}
Brüche = {Bruch(z,n) for z in Zähler for n in Nenner}

for n in Brüche:
    print(n,end=' ')

print('\n')
    
print(Brüche)

# help(Fraction)
```

```python
print(sum(A))              # Alle Zahlen von 1 - 100 = 101 * 50 
print(sum(Brüche))
```
