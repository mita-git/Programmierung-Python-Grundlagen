### Einfaches Indizieren und Slicing

Das Indizieren von Arrays in NumPy ist ein weites Feld, weil es sehr viele Möglichkeiten gibt, eine Teilmenge Ihrer Daten oder einzelne Elemente auszuwählen. Eindimensionale Arrays sind einfach; oberflächlich betrachtet, verhalten sie sich ähnlich wie Python-Listen

```python
import numpy as np
NpArray5 = np.arange(10)
NpArray5
```

```python
NpArray5[5]
```

```python
NpArray5[5:8]
```

```python
NpArray5[5:8] = 42
NpArray5
```

Ein erster wichtiger Unterschied zu Pythons eingebauten Listen besteht darin, dass **Teilbereiche (Slices) von Arrays** sogenannte **Views** auf das ursprüngliche Array sind.

```python
arr_slice = NpArray5[5:8]
arr_slice
```

Wenn ich nun die Werte in arr_slice ändere, spiegeln sich die Änderungen im Original-Array arr wider:

```python
arr_slice[1]=42*2
print("arr_slice",arr_slice)
print("NpArray5",NpArray5)
```

Das »leere« Slice [:] wird allen Werten in einem Array zugewiesen:

```python
arr_slice[:] = 64
print("NpArray5",NpArray5)
```

Falls Sie statt eines Views tatsächlich die Kopie eines Slice eines ndarray haben wollen, müssen Sie das Array explizit kopieren – zum Beispiel so: arr[5:8].copy().

## 4.2 Universelle Funktionen: schnelle elementweise Array-Funktionen

Eine universelle Funktion oder ufunc ist eine Funktion, die auf den Daten in ndarrays elementweise Operationen durchführt.<br>
Diese werden als unäre ufuncs bezeichnet.<br>


```python
npArrX10 = np.arange(10)
npArrX10
```

```python
np.sqrt(npArrX10)
```

Andere, wie add oder maximum, nehmen zwei Arrays entgegen (daher binäre ufuncs) und liefern ein einzelnes Array als Ergebnis zurück:

```python
x = np.random.randn(8)
y = np.random.randn(8)
print('X',x)
print('Y',y)
```

```python
np.maximum(x, y)
# Hier hat numpy.maximum das elementweise Maximum der Elemente in x und y berechnet
```

Es ist zwar nicht üblich, aber eine ufunc kann auch mehrere Arrays zurückgeben. **modf** ist ein Beispiel dafür, eine vektorisierte Version der eingebauten Python-Funktion **divmod**; es liefert die gebrochenen und ganzzahligen Teile eines Gleitkomma Arrays zurück.

```python
NpArr7 = np.random.randn(10) * 10
remainder, whole_part = np.modf( NpArr7 )
print('NpArr7',NpArr7)
print('remainder',remainder)
print('whole_part',whole_part)
```

```python
np.sqrt(NpArr7)
```

## 4.3 Array-orientierte Programmierung mit Arrays
Mit NumPy-Arrays können Sie viele Arten von Datenverarbeitungsaufgaben in Form präziser Array-Ausdrücke schreiben, für die ansonsten Schleifen erforderlich wären. Dieses Ersetzen expliziter Schleifen durch Array-Ausdrücke bezeichnet man gemeinhin als Vektorisierung. Im Allgemeinen sind vektorisierte Array-Operationen um ein oder zwei (oder sogar mehr) Größenordnungen schneller als ihre reinen Python-Gegenstücke. Ihre größte Wirkung zeigen sie bei allen Arten von numerischen Berechnungen.

```python
# Die Funktion np.meshgrid nimmt zwei eindimensionale Arrays entgegen und 
# erzeugt zwei zweidimensionale Matrizen entsprechend den Paaren von (x, y) in den beiden Arrays.

points = np.arange(-5, 5, 0.01) # 1000 equally spaced points from -5 to 5
print('points',points.size)
print('points',points[110:120])
print('points',points[710:720])
print('points min',points.min(),'points max',points.max())
```

```python
xs, ys = np.meshgrid(points, points)

print('points xs, ys',xs.size,ys.size)

print('points xs',xs[100][10:20])
print('points ys',ys[700][10:20])
```

```python
z = np.sqrt(xs ** 2 + ys ** 2)
print('points',z.size)
```

```python
import matplotlib.pyplot as plt

plt.imshow(z); plt.colorbar()
plt.title("Image plot of $\sqrt{x^2 + y^2}$ for a grid of values")  

# Die matplotlib-Funktion imshow zeichnet ein Bild aus dem zweidimensionalen Array von Funktionswerten.
```

```python
plt.imshow(xs); plt.colorbar(); plt.title("Image plot of $xs$ for a grid of values")  
```

```python
plt.imshow(ys); plt.colorbar(); plt.title("Image plot of $ys$ for a grid of values")  
```

### Bedingte Logik als Array-Operationen ausdrücken

Die Funktion **numpy.where** ist eine vektorisierte Version des **ternären Ausdrucks x if condition else y**. 

```python
#  Nehmen wir an, wir hätten zwei Arrays mit Werten

NpArrX = np.array([1.1, 1.2, 1.3, 1.4, 1.5])
NpArrY = np.array([2.1, 2.2, 2.3, 2.4, 2.5])

# und ein boolesches Array 

NpArrCond = np.array([True, False, True, True, False])

# Nun wollen wir einen Wert aus xarr nehmen, wenn der korrespondierende Wert in cond True ist. 
# Andernfalls soll der Wert aus yarr genommen werden.

result = [(x if c else y) 
    for x, y, c in zip(NpArrX, NpArrY, NpArrCond)]
result
```

```python
#  Mit np.where können Sie dies sehr knapp und prägnant schreiben:
result = np.where(NpArrCond, NpArrX, NpArrY)
result
```

```python
#  Umkehrung der cond
result = np.where(NpArrCond == False, NpArrX, NpArrY)
result
```

```python
# Das zweite bzw. dritte Argument zu np.where muss kein Array sein, eines oder beide können Skalare sein.

NpArr4x4 = np.random.randn(4, 4)
NpArr4x4
```

```python
NpArr4x4 > 0
```

```python
 np.where(NpArr4x4 > 0, 'positv', 'negativ')
```

### Mathematische und statistische Methoden

Eine Reihe mathematischer Funktionen, die Statistiken über ein ganzes Array oder über die Daten entlang einer Achse berechnen, sind als Methoden der Array-Klasse verfügbar. Aggregationen (oft auch **Reduktionen** genannt) wie die Summe **sum**, der Mittelwert **mean** und die Standardabweichung **std** können entweder als Methoden der Array-Instanz oder auf oberster Ebene als NumPy-Funktion aufgerufen werden.

```python
NpArr5x4 = np.random.randn(5, 4)
NpArr5x4    
```

```python
NpArr5x4.mean()
```

```python
NpArr5x4.sum()
```

```python
NpArr5x4.std()
```

```python
print('Mittelwerte über Zeilen (axis=0)',NpArr5x4.mean(axis=0))
print('Mittelwerte über Spalten (axis=1)',NpArr5x4.mean(axis=1))
```

### Methoden für boolesche Arrays

während **all** prüft, ob alle Werte True sind:

```python
bools = np.array([False, False, False, True])
print('Any von Bools:',bools.any())
print('All von Bools:',bools.all())
nums = np.array([1, 2, 3, 4])
print('All von nums:',nums.all())
```
