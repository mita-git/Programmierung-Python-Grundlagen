e) Berechnen Sie ein **Dictionary wdict**, dass die Wortlängen als Schlüssel verwendet und zu jeder Wortlänge als Wert die Anzahl der Worte mit genau dieser Länge speichert.<br>
f) Verwenden Sie die Bibliothek **Matplotlib** um die Verteilung der Wortlängen als Funktion zu plotten.<br>
Auf der **x-Achse sollen die Wortlängen** (in Zeichen) aufgetragen sein, auf der **y-Achse die Anzahl der Worte** mit genau x Zeichen. Sie können die Funktion als Kurve oder als Scatter-Plot darstellen.<br>


```python
## Quelle: https://netzmafia.ee.hm.edu/software/wordlists/deutsch.txt

FileHandler = open('deutsch.txt', 'r')

Worte = list()

for line in FileHandler:
    Worte.append(line.strip())

FileHandler.close()

print("Anzahl der Worte:",len(Worte))

wdict = dict()
for Wort in Worte:
    WortLaengen = len(Wort)
    if WortLaengen not in wdict:
        wdict[WortLaengen] = 1
    else:
        wdict[WortLaengen] += 1
print("Vorkommen der Länge der Worte")
print(wdict)

import matplotlib.pyplot as plt 

x = list(wdict.keys())
y = list(wdict.values())


plt.bar(x,y)
plt.title('Vorkommen der Länge der Worte in deutsch.txt '+str(len(Worte))+' Worte')
plt.xlabel('Länge der Worte')
plt.ylabel('Vorkommen')
plt.show()
```
