```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import math
```

```python
x = np.linspace(0, 2 * math.pi, 100)
y1 = np.cos(x)
y2 = np.sin(x)

plt.plot(x, y1)
plt.plot(x, y2)

line1, line2 = plt.gca().lines
plt.setp(line2, linewidth=5)
plt.setp(line1, color='m', linewidth=5)
plt.setp(line1, linestyle='--')
plt.setp(line2, linestyle='-.')
```

```python
x = np.linspace(0, 2 * math.pi, 30)
y1 = np.cos(x)
y2 = np.sin(x)

plt.plot(x, y1)   # cosinus
plt.plot(x, y2)   # sinus

Cosinus, Sinus = plt.gca().lines

# Bei Bedarf lassen sich die Eigenschaften von Symbolen im Detail festlegen. 
# Das folgende Beispiel erzeugt auf der Linie 2 quadratische blaue Symbole mit einem roten Rand von 3 Punkten Breite.

plt.setp(Cosinus, marker='s', markerfacecolor='r', markeredgewidth=2, markeredgecolor='b', label='Cosinus')
plt.setp(Sinus, marker='s', markerfacecolor='b', markeredgewidth=2, markeredgecolor='r', label='Sinus')


plt.xlabel('$Zeit \ in \ sec$')
plt.ylabel(r'$\cos(\omega_t), \sin(\omega_t)$')

plt.legend()

plt.draw()
```

```python
plt.cla()
plt.plot(x, y1, 'o-', label='Cosinus')
plt.plot(x, y2, 's-', label='Sinus')

# plt.legend(loc='upper right')
plt.legend(bbox_to_anchor=(1, 1), loc='upper left', borderaxespad=1)
```

```python
plt.cla()
plt.plot(x, y1, 'o-', label='Cosinus')
plt.plot(x, y2, 's-', label='Sinus')


plt.xticks(np.pi*np.arange(0, 4), ('0', r'$\pi$', r'$2\pi$', r'$3\pi$'),size='x-large', color='r')

plt.grid(which='both')

plt.xlim(math.pi - 1 , math.pi + 1 )
```

```python
plt.cla()
plt.plot(x, y1, 'o-', label='Cosinus')
plt.plot(x, y2, 's-', label='Sinus')

plt.yscale('log')
plt.grid(which='both')
```

## 3D Funktionen anzeigen

```python
import warnings
warnings.filterwarnings('ignore') # setting ignore as a parameter

from mpl_toolkits.mplot3d import Axes3D

x, y = np.mgrid[-3:3:30j, -3:3:30j]
z = (x**2+y**3)*np.exp(-x**2-y**2)
cmap = 'coolwarm'

fig = plt.figure()
ax = fig.gca(projection='3d')

ax.plot_surface(x, y, z, rstride=1, cstride=1, cmap=cmap, alpha=0.5)
cset = ax.contourf(x, y, z, zdir='z', offset=-0.8, cmap=cmap)
ax.set_xlabel('$x$', size='xx-large')
ax.set_ylabel('$y$', size='xx-large')
ax.set_zlabel('$z$', size='xx-large')
ax.set_zlim(-0.8, 0.5)

plt.draw()
```
